# **SEGUNDA ENTREGA**

2.1. LIMPIEZA DEL DATASET


```python
# Drop the Student_ID column
df = df.drop(columns=["Student_ID"])
df = df.drop(columns=["Department"])
df["Gender"] = df["Gender"].map({"Male": 0, "Female": 1})

df["Depression"] = df["Depression"].map({False: 0, True: 1})
print(df)
```

2.2 PARTICIONADO 80/20 + SHUFFLE

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler


X = df.values[:, :-1]
y = df.values[:, -1]


X_small, _, y_small, _ = train_test_split(
    X, y,
    train_size=20000,
    stratify=y,
    random_state=42
)


X_train, X_test, y_train, y_test = train_test_split(
    X_small, y_small,
    test_size=0.2,
    random_state=1,
    shuffle=True,
    stratify=y_small
)


scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

3.1 DECISION TREE

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

est = DecisionTreeClassifier(max_depth=7, class_weight="balanced")
est.fit(X_train,y_train)
print(accuracy_score(est.predict(X_test), y_test))
acc_dt = accuracy_score(est.predict(X_test), y_test)
```

3.2 RANDOM FOREST

```python
from sklearn.ensemble import RandomForestClassifier
est = RandomForestClassifier(n_estimators = 100, class_weight="balanced")
est.fit(X_train,y_train)
print(accuracy_score(est.predict(X_test), y_test))
acc_rf = accuracy_score(est.predict(X_test), y_test)
```

3.3 SVM

```python
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)


est = SVC(kernel="poly", class_weight="balanced")
est.fit(X_train_scaled, y_train)
print(accuracy_score(est.predict(X_test_scaled), y_test))
acc_svm_poly = accuracy_score(est.predict(X_test_scaled), y_test)


est = SVC(kernel="linear", class_weight="balanced")
est.fit(X_train_scaled, y_train)
print(accuracy_score(est.predict(X_test_scaled), y_test))
acc_svm_linear = accuracy_score(est.predict(X_test_scaled), y_test)


est = SVC(kernel="rbf", class_weight="balanced")
est.fit(X_train_scaled, y_train)
print(accuracy_score(est.predict(X_test_scaled), y_test))
acc_svm_rbf = accuracy_score(est.predict(X_test_scaled), y_test)
```

4.1 CROSSVALIDATION DT

```python
def tpr(est,X,y):
    p = est.predict(X)
    return np.mean (p[y==1] == y[y==1])

def tnr(est,X,y):
    p = est.predict(X)
    return np.mean(p[y==0] == y[y==0])
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import KFold
from sklearn.metrics import make_scorer
from sklearn.model_selection import cross_val_score, StratifiedKFold

est = DecisionTreeClassifier(max_depth=7,class_weight="balanced", random_state=42)
cv = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

sdt1 = cross_val_score(est, X_small, y_small, cv=cv, scoring=make_scorer(accuracy_score))
sdt2 = cross_val_score(est, X_small, y_small, cv=cv, scoring=tpr)
sdt3 = cross_val_score(est, X_small, y_small, cv=cv, scoring=tnr)

print("accuracy %.3f (+/- %.5f)"%(np.mean(sdt1), np.std(sdt1)))
print("tpr      %.3f (+/- %.5f)"%(np.mean(sdt2), np.std(sdt2)))
print("tnr      %.3f (+/- %.5f)"%(np.mean(sdt3), np.std(sdt3)))
```

4.2 CROSSVALIDATION RF

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score, StratifiedKFold

est = RandomForestClassifier(n_estimators=100, class_weight="balanced",random_state=42)

cv = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)


srf1 = cross_val_score(est, X_small, y_small, cv=cv, scoring=make_scorer(accuracy_score))
srf2 = cross_val_score(est, X_small, y_small, cv=cv, scoring=tpr)
srf3 = cross_val_score(est,X_small, y_small, cv=cv, scoring=tnr)

print("accuracy %.3f (+/- %.5f)"%(np.mean(srf1), np.std(srf1)))
print("tpr      %.3f (+/- %.5f)"%(np.mean(srf2), np.std(srf2)))
print("tnr      %.3f (+/- %.5f)"%(np.mean(srf3), np.std(srf3)))
```

4.3 CROSSVALIDATION SVM

```python
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.metrics import make_scorer
from sklearn.metrics import accuracy_score

est = SVC(kernel="linear",class_weight="balanced", random_state=42 )
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

svm1 = cross_val_score(est, X_small, y_small, cv=cv, scoring=make_scorer(accuracy_score))
print("accuracy %.3f (+/- %.5f)"%(np.mean(svm1), np.std(svm1)))
svm2 = cross_val_score(est,X_small, y_small, cv=cv, scoring=tpr)
print("tpr      %.3f (+/- %.5f)"%(np.mean(svm2), np.std(svm2)))
svm3 = cross_val_score(est, X_small, y_small, cv=cv, scoring=tnr)
print("tnr      %.3f (+/- %.5f)"%(np.mean(svm3), np.std(svm3)))
```

3.0 StandardScaler AL DATASET

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler


scaler = StandardScaler()
scaled_df = scaler.fit_transform(df)
sclar_df = pd.DataFrame(scaled_df, columns=df.columns)
print(sclar_df)
```

3.1 REDUCCIÓN DIMENSIONAL

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2 , whiten = True )
pca_features = pca.fit_transform(sclar_df)
print(pca_features)
```

```python
print(pca.explained_variance_ratio_)
print(sum(pca.explained_variance_ratio_))
```

```python
from sklearn.cluster import KMeans

inertia = []

for k in range(1,10):
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(scaled_df)
    inertia.append(km.inertia_)

plt.plot(range(1,10), inertia, marker='o')
plt.xlabel("Número de clusters")
plt.ylabel("Inercia")
plt.show()
```

3.2 APLICAR KMEANS

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2, random_state=21)
kmeans_clusters = kmeans.fit_predict(pca_features)
centroids = kmeans.cluster_centers_
```

3.3 APLICAR DBSCAN

```python
from sklearn.cluster import DBSCAN

from sklearn.metrics import silhouette_score


parametros = [
    (0.02, 3),
    (0.03, 4),
    (0.05, 5),
    (0.1 , 2),
    (0.2 , 30)
]

for eps, min_samples in parametros:
    dbscan = DBSCAN(eps=eps, min_samples=min_samples)
    labels = dbscan.fit_predict(pca_features)
    score = silhouette_score(pca_features, labels)
    print(f"eps={eps}, min_samples={min_samples}, silhouette={score:.3f}")
```

```python
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps=0.2, min_samples=30)
DBSCAN_Cluster = dbscan.fit_predict(pca_features)
```

```python
y = df["Depression"]
plt.figure(figsize=(8,6))

plt.scatter(
    pca_features[:,0],
    pca_features[:,1],
    c=y,
    cmap='coolwarm',
    alpha = 0.5
)

plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA coloreado por depresión")
plt.show()
```

3.4 GRAFICAR RESULTADOS

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(15,5))


plt.subplot(1,3,1)
plt.scatter(pca_features[:,0], pca_features[:,1], c='gray', s=30)
plt.title("Datos originales (PCA)")
plt.xlabel("PC1")
plt.ylabel("PC2")


plt.subplot(1,3,2)
plt.scatter(pca_features[:,0], pca_features[:,1], c=kmeans_clusters, cmap='viridis', s=30)
plt.scatter(centroids[:,0], centroids[:,1], c='red', marker='X', s=200, label='Centroides')
plt.title("Clusters KMeans")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.legend()


plt.subplot(1,3,3)
plt.scatter(pca_features[:,0], pca_features[:,1], c=DBSCAN_Cluster, cmap='plasma', s=30)
plt.title("Clusters DBSCAN")
plt.xlabel("PC1")
plt.ylabel("PC2")

plt.tight_layout()
plt.show()
```
